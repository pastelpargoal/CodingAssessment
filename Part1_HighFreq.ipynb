# 1. Survey Duration

## a. Calculate the average and median values of time spent surveying, considering only completed surveys.

```python
#Subsetting to those that only completed the surveys
surveydata1= surveydata[surveydata.survey_complete==1]
surveydata1.head()
```

```python
# Avg and median
surveyAvg= float(surveydata1.surveytime.mean())
surveyMedian = surveydata1.surveytime.median()
# Displaying the results
surveyAvg
```
